#Handling Outliers

```python
df['campaign'].describe()
```

```python
upper_limit = df['campaign'].quantile(0.99)
lower_limit = df['campaign'].quantile(0.01)
```

```python
df['campaign'] = np.where(df['campaign'] >= upper_limit,
        upper_limit,
        np.where(df['campaign'] <= lower_limit,
        lower_limit,
        df['campaign']))
```

```python
sns.boxplot(df['campaign'])
```

```python
import numpy as np

df['balance'] = np.sign(df['balance']) * np.log1p(np.abs(df['balance']))
```

```python
sns.boxplot(df['balance'])
```

```python
df['pdays'].describe()
```

```python
df['prev_contacted'] = (df['pdays'] != -1).astype(int)


```

```python
df['pdays'] = df['pdays'].replace(-1, 0)
```

```python
df['pdays'].min()
```

```python


bins = [-0.1, 0, 30, 90, 365, np.inf]
labels = ['never', 'recent', 'medium', 'old', 'very_old']

df['pdays'] = pd.cut(
    df['pdays'],
    bins=bins,
    labels=labels,
    include_lowest=True
)
```

```python
df['pdays'].unique()
```

```python
df['pdays'].isnull().sum()
```

```python
sns.countplot(x='pdays', data=df)
```

```python
cat_cols = ['job', 'marital', 'education', 'housing', 'loan', 'contact',
       'poutcome','pdays']
```

```python
num_cols = ['age','balance', 'campaign', 'previous','prev_contacted']
```

```python
df['pdays']
```

```python
trf = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(drop='first', sparse_output=False), cat_cols),
        ('num', StandardScaler(), num_cols)
    ]
)
```

```python
df
```

```python
X_transformed = trf.fit_transform(df)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.95, random_state=42)
X_pca = pca.fit_transform(X_transformed)
```

```python
X_pca.shape
```

```python
pca.explained_variance_ratio_.sum()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA  95% Variance Retained")
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []

for k in range(2, 11):
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_pca)
    inertia.append(km.inertia_)

plt.figure()
plt.plot(range(2, 11), inertia)
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

for k in [4, 5, 6, 7]:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X_pca)
    score = silhouette_score(X_pca, labels)
    print(f"k={k}, silhouette={score:.3f}")
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=6,
    random_state=42,
    n_init=10
)

df['cluster'] = kmeans.fit_predict(X_pca)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7, 5))
plt.scatter(
    X_pca[:, 0],
    X_pca[:, 1],
    c=df['cluster'],
    cmap='tab10',
    alpha=0.7
)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("Customer Segments (K-Means on PCA space)")
plt.colorbar(label='Cluster')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(
    X_pca[:, 0],
    X_pca[:, 1],
    X_pca[:, 2],
    c=df['cluster'],
    cmap='tab10',
    alpha=0.7
)

ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
ax.set_title("3D PCA Cluster Visualization")
plt.show()
```

#Interpretation Of The Clusters

 Cluster 0  Over-Contacted New Customers (Niche)

Who they are: Middle-aged professionals with moderate balances and no prior contact history.
Behavior: Extremely high campaign contact despite being first-time targets.
Insight: Inefficient targeting with excessive outreach.
Action: Reduce contact frequency; refine targeting rules.

 Cluster 1  Young New Entrants (Core Segment)

Who they are: Young, working professionals with good balances and no previous campaign exposure.
Behavior: Low campaign contact, new to marketing efforts.
Insight: Large, untapped growth segment.
Action: Onboarding campaigns and entry-level financial products.

 Cluster 2  Low-Balance Low-Engagement Customers

Who they are: Middle-aged professionals with very low balances and minimal engagement.
Behavior: Limited past contact and low responsiveness.
Insight: Low profitability segment.
Action: Cost-efficient digital outreach; avoid aggressive marketing.

 Cluster 3  Highly Contacted Repeat Customers (Very Niche)

Who they are: Previously contacted customers with long engagement history.
   Behavior: Extremely high past contact counts but poor campaign outcomes.
Insight: Repeated unsuccessful targeting.
Action: Pause generic campaigns; switch to personalized advisory approaches.

 Cluster 4  Stable Previously Engaged Customers (Large)

Who they are: Mature, financially stable customers with prior campaign exposure.
Behavior: Low current contact and moderate past engagement.
Insight: High-efficiency, low-maintenance segment.
Action: Retention and loyalty-based strategies.

 Cluster 5  Older High-Value New Customers (Large)

Who they are: Older customers with the highest balances and no campaign history.
Behavior: Financially strong but under-engaged.
Insight: High-value segment with strong upsell potential.
Action: Premium, trust-focused marketing and long-term products.

#The six clusters capture distinct customer behaviors ranging from untapped high-value clients to over-contacted low-response segments, enabling targeted and efficient marketing strategies.
